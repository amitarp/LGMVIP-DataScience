## Importing the necessary files

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

### Reading the file

```python
#reading the image
image=cv2.imread('patronus.jpg')
```

### Resizing the image according to our requirements

```python
image = cv2.resize(image, (600, 500))
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## Converting the image to grayscale

```python
grayfilter=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grayfilter))
```

### Inverting the image

```python
invert=cv2.bitwise_not(grayfilter)
display(Image.fromarray(invert))
```

### Bluring the image

```python
blurimage=cv2.GaussianBlur(invert,(5,5),0)
display(Image.fromarray(blurimage))
```

### Inverting the image to its original form

```python
invertedblur=cv2.bitwise_not(blurimage)
display(Image.fromarray(invertedblur))
```

### Final image

```python
sketch_filter=cv2.divide(grayfilter,invertedblur,scale=250.0)
display(Image.fromarray(sketch_filter))
```
