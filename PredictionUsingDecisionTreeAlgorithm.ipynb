### Importing all the libraries

### Data PreProcessing

```python
X=data.iloc[:,[0,1,2,3]]
Y=data.Species
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

### Data split into training and test set

```python
#Spliting the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

### PreProcessing the Data

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Predicting the data using Decision Tree

```python
# Making the instant for Decision Tree Clasifier
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
```

```python
pd.DataFrame(data.Species).value_counts()
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
#For training set, accuracy is 1 while for test set, accuracy is .96
```

```python
#Ploting the Decision Tree
rcParams['figure.figsize'] = 80, 50
plot_tree(clf,precision=5,rounded=True,filled=True);
```
